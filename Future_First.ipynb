```python
#Removing the price column
column=stock_df.pop("Price")
column
```

```python
#Adding the price column at the 5th postion and renaming it as Close
stock_df.insert(4,'Close',column)
stock_df
```

```python
#Changing the datatype of Date from object to Datetime
stock_df['Date']=pd.to_datetime(stock_df['Date'])
stock_df
```

```python
#Setting the Date as Index
stock_df = stock_df.set_index(pd.DatetimeIndex(stock_df['Date'].values))
```

```python
#Sorting the dataset in the  ascending order of Date
stock_df.sort_values(by='Date', inplace = True) 
stock_df
```

```python
#Calculating the True range 
high_low = stock_df['High'] - stock_df['Low']
high_cp = np.abs(stock_df['High'] - stock_df['Close'].shift())
low_cp = np.abs(stock_df['Low'] - stock_df['Close'].shift())
```

```python
#Concating the True range values to calculate the ATR
df=pd.concat([high_low,high_cp,low_cp],axis=1)
true_range = np.max(df, axis=1)
true_range
```

```python
# Getting the ATR value on the rolling of 14
avg_true_range = true_range.rolling(14).mean()
avg_true_range
```

```python
#Adding the ATR column in the dataset
stock_df['atr'] = avg_true_range
stock_df
```

```python
#define the window for moving standard deviation
window = input("Enter the window size for moving standard deviation: ")
```

```python
#calculate the moving standard deviation
stock_df['Std_Dev_Close'] = stock_df['Close'].rolling(window=int(window)).std()
stock_df['Std_Dev_ATR'] = stock_df['atr'].rolling(window=int(window)).std()
stock_df.head(20)
```

```python
#Importing the plotly for visualization
import plotly.graph_objects as go
```

```python
fig= go.Figure(
    data = [
            go.Candlestick(
                x = stock_df.index,
                low = stock_df['Low'],
                high = stock_df['High'],
                close = stock_df['Close'],
                open = stock_df['Open'],
                increasing_line_color = 'red',
                decreasing_line_color = 'black',
                name = 'Daily Price'              
            )
    ]
)

# Add the ATR chart on a different vertical axis
fig.add_trace(go.Scatter(x=stock_df.index, y=stock_df['atr'], name="ATR", yaxis='y2'))

# Add input box for windowsize
fig.update_layout(
    updatemenus=[
        go.layout.Updatemenu(
            type='dropdown',
            buttons=list([
                dict(
                    args=[{"y": [stock_df['Close'],stock_df['atr'].rolling(window=i).std()]}],
                    label=str(i),
                    method='update'
                ) for i in [0,10,20,30,50,100,200,250]
            ]),
            direction = "down",
            pad = {"r": 10, "t": 10},
            showactive = True,
            x = 0.1,
            xanchor = "left",
            y = 1.1,
            yanchor = "top"
        ),
    ])


# Adding the moving standard deviation on an chart
fig.add_trace(go.Scatter(x=stock_df.index, y=stock_df['Std_Dev_Close'], name="Standard Deviation Close", yaxis='y3'))
fig.add_trace(go.Scatter(x=stock_df.index, y=stock_df['Std_Dev_ATR'], name="Standard Deviation ATR", yaxis='y4'))

fig.update_layout(
    title = 'German 10 Year Bund Futures ',
    yaxis_title = 'Euros(â‚¬)',
    yaxis2=dict(title='ATR', overlaying='y', side='right'),
    yaxis3=dict(overlaying='y', side='left'),
    yaxis4=dict(overlaying='y', side='left'),
    xaxis_title = 'Date Range'    
)

fig.show()
     
```
